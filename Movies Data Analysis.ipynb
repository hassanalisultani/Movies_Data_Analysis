```python
# Import libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

from matplotlib.pyplot import figure

plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)

pd.options.mode.chained_assignment = None


df = pd.read_csv('E:\Data Analysis Project\Movies_Data_Analysis\movies.csv')
```

```python
# pd.set_option('display.max.rows', None) # "To show all rows"
df
```

```python
# Let's see if we have any missing data

for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing_pct*100)))
```

```python
# Correlation Matrix of btweeen all Numeric Columns

df.corr(method='pearson') # Pearson is a default method for the function 'dataframe.corr()'
```

```python
df.corr('kendall')
```

```python
df.corr('spearman')
```

```python
correlation_matrix = df.corr() # By default 'Pearson' selected
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix of numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

df.apply(lambda x:x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x:x.factorize()[0]).corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for Movies')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
correlation_mat = df.apply(lambda x:x.factorize()[0]).corr()
corr_pairs = correlation_mat.unstack()
print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind='quicksort')
print(sorted_pairs)
```

```python
# Take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs)>0.5]
print(strong_pairs)
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[['gross']].sum()
CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending=False)[:15]
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64')
CompanyGrossSumSorted
```

```python
df.groupby(['company', 'year'])[[('gross')]].sum()
```

```python
CompanyGrossSum = df.groupby(['company', 'year'])[['gross']].sum()
CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross', 'company', 'year'], ascending=False)[:15]
CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64')
CompanyGrossSumSorted
```

```python
df_numerized = df

for col in df_numerized.columns:
    if (df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes

df_numerized
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for Movies')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```
